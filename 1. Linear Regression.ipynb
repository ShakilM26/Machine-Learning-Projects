## In Linear Regression, 2 variables are related through an equation, where exponent(power) of both these variables is 1.

* Mathematically a linear relationship represents a straight line when plotted as a graph

### Non-linear relationship where the exponent of any variable is not equal to 1 and creates a curve.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
data.head(2)
```

```python
sns.regplot(x='total_bill', y='tip', data=data)
plt.show()
```
